Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

radianes = grados*(pi/180)

```python
import math
grados = float(input("Ingresa los grados a pasar a radianes:"))
radianes = round(grados*(math.pi/180), 4)
print(f"{grados} grados representan {radianes} radianes")
```

Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`

```python
base = float(input("Ingrese la base del paralelogramo (en metros):"))
altura = float(input("Ingrese la altura del paralelogramo (en metros):"))
area = round(base * altura, 2)
print(f"El area del paralelogramo es de {area} metros cuadrados")
```



## Ejercicio 4
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`. En este caso debe ser capaz de admitir valores reales, con decimales, para base y altura en español. (PISTA: Recuerda que tendrás que hacer uso del método replace)

```python
base = float(input("Ingrese la base del paralelogramo (en metros):").replace(",", "."))
altura = float(input("Ingrese la altura del paralelogramo (en metros):").replace(",", "."))
area = round(base * altura, 2)
print(f"El area del paralelogramo es de {area} metros cuadrados")
```

Complete el siguiente código para solicitar al usuario su nombre y su edad. Luego, calcule en qué año nació el usuario y muestre el resultado.

```python
nombre = input("¿Cuál es tu nombre? ")
edad = int(input("¿Cuál es tu edad? "))
cumplido = input("¿Ya cumpliste años? (y/n)")
año_actual = 2025  # Cambiar según el año en curso
while cumplido != "y" and cumplido != "n":
    cumplido = input("Ese caracter no existe. Por favor ingresa y/n:")
if cumplido == "y":
    año_nacimiento = año_actual - edad
else:
    año_nacimiento = año_actual - edad - 1

print(f"{nombre}, naciste en el año {año_nacimiento}.")
```
